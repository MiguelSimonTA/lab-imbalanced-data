We will try to predict variable Churn using a logistic regression on variables tenure, SeniorCitizen,MonthlyCharges.

```python
customer_churn_dropped = customer_churn.drop(['customerID', 'gender', 'Partner', 'Dependents',
        'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'TotalCharges'], axis =1)
```

```python
customer_churn_dropped
```

Extract the target variable.
Extract the independent variables and scale them. (I chose not to scale it, don't know how to do it..)
Build the logistic regression model.

```python
X = customer_churn_dropped.drop('Churn',axis = 1)
y = customer_churn_dropped['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.2)

LR = LogisticRegression()
LR.fit(X_train, y_train)

LR.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report


pred = LR.predict(X_test)
pred


print("precision: ",precision_score(y_test,pred, pos_label='Yes'))
print("recall: ",recall_score(y_test,pred, pos_label='Yes'))
print("f1: ",f1_score(y_test,pred, pos_label='Yes'))


print(classification_report(y_test, pred)) 
```

```python

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

evaluate the model

Our precison is approximately 70% (we were right 70% of time), the racall is 47%(from all the churn we identified 47%) and the f1 score is 56%, it is considered average.

Even a simple model will give us more than 70% accuracy. Why?
Because the data is imbalanced and number of customers staying is not high.


# SMOTE

Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest neighbors that adds new points between existing points. Apply imblearn.over_sampling.SMOTE to the dataset. Build and evaluate the logistic regression model. Is it there any improvement?


```python
from imblearn.over_sampling import SMOTE


# a bit of magic, you can find documentation here: https://imbalanced-learn.readthedocs.io/en/stable/generated/imblearn.over_sampling.SMOTE.html
# by deafult, takes a 5-neightbour KNN to build a new point
sm = SMOTE(random_state =0,sampling_strategy=1) # one means that I want a ratio of 1 between majority and abudance classe. # 0.5 Means my minority class will be half has big as my majority class
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train,y_train)
```

```python
train_smote = pd.concat([X_train_SMOTE,y_train_SMOTE], axis = 1)
train_smote
train_smote.sum() # 402 with 1 and 402 with 0
```

```python
# yet another small improvement, but bear in mind that we saved 12 hypothetical people with
# these "small improvements"
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(X_train_SMOTE, y_train_SMOTE)
pred = LR.predict(X_test)

print("precision: ",precision_score(y_test,pred, pos_label='Yes'))
print("recall: ",recall_score(y_test,pred, pos_label='Yes'))
print("f1: ",f1_score(y_test,pred, pos_label='Yes'))

print(classification_report(y_test, pred))
```

```python
confusion_matrix(y_test,pred)
```

The model improved not in the accuracy, but when we made the data more balanced the Recall improved, and we have less False negatives, that was the most important not to lose many more customers. In this case we are identifying the customers that we will lose better, although we are less correct about the ones that we catch.
