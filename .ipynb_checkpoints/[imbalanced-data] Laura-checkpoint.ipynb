```python
# Libaries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.utils import resample
```

```python
#### Preparations for the data
# Data
df = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\07_Logarithmic_regression\lab-imbalanced-data\files_for_lab\customer_churn.csv')
def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df)

df
# 7.043 x 21
```

```python
# Taget: churn
# independant variables: 'tenure', 'seniorcitizen','onthlycharges'

# Checking for NaN values
sum_na = df.isna().sum()
print (sum_na)
```

```python
# Exploring data
data_types = df.dtypes
print (data_types)

print(df['churn'].unique())
```

```python
# Target value --> churn
df['churn'] = df['churn'].replace({'Yes': 1, 'No': 0})
df
```

```python
# Assessing distribution of data in terms of the target value:
df.groupby('churn').count()
#--> unbalanced data! 0 = 5.174 vs 1 = 1869
```

```python
model_df = df[['tenure', 'seniorcitizen','monthlycharges']]
model_df
```

```python
# Scale/Normalise the independant variables
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalised_data = scaler.fit_transform(model_df)
normalised_data = pd.DataFrame(normalised_data, columns = model_df.columns)
normalised_data
```

```python
# Regression model
#1. Split the data

X = model_df
Y = df['churn']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#2. Create the model:

LR = LogisticRegression()
LR.fit(X_train, Y_train)
LR.score(X_test, Y_test)

# -> Accuracy of 76 % that I can predict the satisfaction state of a customer

# Other evaluation metrics:

predictions = LR.predict(X_test) # Here prediciting Y 

print('Accuracy is: ',  accuracy_score(Y_test, predictions))
print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))
```

```python
#3. confusion matrix
confusion_matrix (Y_test, predictions)

'''
Interpretation:
- 80 customers have been falsely identified as likely to return
- 194 customers have been falsely identified as unlikely to return
--> However, even though the accuracy is high (>80%), the results are likelty not representative, because of the inbalance
    in the dataset
    
--> Accuracy is so high because the model likely predicts the majority class correctly
'''
```

```python
### Optimising the model
# Undersampling

# 1. Concate the train data again
train = pd.concat([X_train, Y_train], axis = 1)

# 2. Seperate minority/majority class
yes_churn = train[train['churn'] == 1]
no_churn = train[train['churn'] == 0]

display(yes_churn.shape) #1496 & 4
display(no_churn.shape) #4138 & 4

#3. Removing rows

no_churn_undersample = resample(no_churn,
                                   replace = False,
                                   n_samples = len(yes_churn),
                                   random_state = 0
                                  )

display(no_churn_undersample.shape) #1496 x 4
display(yes_churn.shape) #1496 x 4
```

```python
#  Now the model
train_undersample = pd.concat ([no_churn_undersample, yes_churn], axis = 0)

# X-Y-Split
X_train_under = train_undersample.drop('churn', axis =1)
Y_train_under = train_undersample['churn']

# Build model
LR = LogisticRegression(max_iter = 1000)
LR.fit(X_train_under, Y_train_under)

# Predictions for test data
predictions = LR.predict(X_test) # Here prediciting Y 

print('Accuary: ', LR.score(X_test, Y_test))

print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))
```

```python
confusion_matrix (Y_test, predictions)

'''
Interpretation:
- 274 customers have been falsely identified as likely to return (before 80)
- 85 customers have been falsely identified as unlikely to return (before 194)
--> With undersampling the false negative cases have drastically decreased (recall gained 30%); Accuracy is still over >70%;
    However, precision has decreased, leaving more false posiive predictions
'''
```

```python
### Optimising the model
# Oversampling

# 1. Concate the train data again
train = pd.concat([X_train, Y_train], axis = 1)

# 2. Seperate minority/majority class
yes_churn = train[train['churn'] == 1]
no_churn = train[train['churn'] == 0]

display(yes_churn.shape) #1496 & 4
display(no_churn.shape) #4138 & 4

#3. Adding rows

yes_churn_oversample = resample(yes_churn,
                                   replace = True,
                                   n_samples = len(no_churn),
                                   random_state = 0 
                                  )

display(yes_churn_oversample.shape) #4138 x 4
display(no_churn.shape) #4138 x 4
```

```python
#  Now the model
train_oversample = pd.concat ([yes_churn_oversample, no_churn], axis = 0)

# X-Y-Split
X_train_under = train_oversample.drop('churn', axis =1)
Y_train_under = train_oversample['churn']

# Build model
LR = LogisticRegression(max_iter = 1000)
LR.fit(X_train_under, Y_train_under)

# Predictions for test data
predictions = LR.predict(X_test) # Here prediciting Y 

print('Accuary: ', LR.score(X_test, Y_test))

print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))
```

```python
confusion_matrix (Y_test, predictions)

'''
Interpretation:
- 272 customers have been falsely identified as likely to return (before 80)
- 86 customers have been falsely identified as unlikely to return (before 194)
--> With oversampling the false negative cases have drastically decreased (recall gained 30%); Accuracy is still over >70%;
--> In comparising with undersampling, 1 more case has been identified as false negative and 2 more as false positive 
'''
```

```python
### Optimising the model
# Oversampling - SMOTE

from imblearn.over_sampling import SMOTE

'''
By default, SMOTE uses 5 nearest neighbour points
'''

sm = SMOTE(random_state = 0, sampling_strategy = 1.0)
X_train_SMOTE, Y_train_SMOTE = sm.fit_resample(X_train, Y_train)
X_train_SMOTE

#  The model
train_oversample = pd.concat([X_train_SMOTE, Y_train_SMOTE], axis = 0)

# Build model
LR = LogisticRegression(max_iter = 1000)
LR.fit(X_train_SMOTE, Y_train_SMOTE)


# Predictions for test data
predictions = LR.predict(X_test)

print('Accuary: ', LR.score(X_test, Y_test))

print('Precision is: ', precision_score(Y_test, predictions))
print('Recall is: ',  recall_score(Y_test, predictions))
print('F1 is: ',  f1_score(Y_test, predictions))

print(classification_report(Y_test, predictions))
```

```python
confusion_matrix (Y_test, predictions)
```
